## Cargamos nuestro dataset de VICTIMAS

```python
df_victimas = pd.read_csv(r'DataBases\DataBases_seed\victimas.csv', encoding='latin1')
df_victimas
```

### PRIMER ANALISIS DE LOS DATOS

```python
df_victimas.info()
```

```python
#Analizamos datos faltantes
null_counts = df_victimas.isnull().sum()
print(null_counts)
```

```python
# Observamos que los duplicados corresponen a los accidentes donde hubieron mas de una victima
duplicated_data = df_victimas[df_victimas.duplicated(subset='ID_hecho')]
print(duplicated_data)
```

```python
# Vemos cuantos registros sin dato hay

sd_counts = (df_victimas == 'SD').sum()
print(sd_counts)
```

```python
#Eliminamos columnas que no son necesarias ya que tenemos los mismos datos en la tabla anterior

df_victimas2 = df_victimas.drop(columns=['FECHA', 'AAAA', 'MM', 'DD', 'VICTIMA', 'FECHA_FALLECIMIENTO'])
df_victimas2
```

```python
# Vemos que hay 3 filas con 3 datos sd , pero decidimos dejarlo de todas maneras y no eliminarlas aunque tengan mas del 75% de los datos faltantes 
row_sd = df_victimas2[df_victimas2['SEXO'] == 'SD']
print(row_sd)
```

### TIPO DE DATO

```python
df_victimas2.info()
```

Cambiar columna edad de object a int

```python
# Reemplaza 'SD' con un valor temporal, por ejemplo, -1
df_victimas2['EDAD'] = df_victimas2['EDAD'].replace('SD', -1)

# Convierte la columna a int
df_victimas2['EDAD'] = df_victimas2['EDAD'].astype(int)
```

Imputación de valores ´SD´ en las edades con la media de las edades 

```python
# Calculamos la media de las edades, excluyendo -1
media_edad = round(df_victimas2[df_victimas2['EDAD'] != -1]['EDAD'].mean(),0)
media_edad
```

```python
# Reemplazamos -1 con la media de las edades
df_victimas2['EDAD'] = df_victimas2['EDAD'].replace(-1, media_edad)
```

### NORMALIZAR DATOS

```python
# Cambiamos nombre de columnas para normalizar datos y unificar con la primera tabla

df_victimas2.rename(columns={'ID_hecho': 'id_hecho'}, inplace=True)
df_victimas2.rename(columns={'ROL': 'rol_victima'}, inplace=True)
df_victimas2.rename(columns={'SEXO': 'sexo_victima'}, inplace=True)
df_victimas2.rename(columns={'EDAD': 'edad_victima'}, inplace=True)
```
