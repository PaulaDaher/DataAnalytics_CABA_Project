## LOCALES BAILABLES Y COMUNAS:

Con el fin de continuar este análisis exploratorio de los datos, importo una base de datos de locales bailables y le hago un pequeño ETL para poder cruzaar datos.  

```python
bailables = pd.read_csv(r'DataBases\DataBases_complementarias\registro-historico-acumulado-de-locales-bailables.csv', encoding='latin1')
bailables
```

```python
bailables.drop(['piso', 'codigo_postal','codigo_postal_argentino','fecha_disposicion_de_rechazo', 'fecha_envio', 'notificado', 'numero_dips', 'numero_disposicion_de_recalculo_capacidad', 'numero_disp_de_rechazo', 'numero_expediente', 'numero_registro', 'estado_observaciones', 'fecha_de_entrega_nueva_revalida', 'vencimiento', 'fecha_de_notificacion', 'fecha_disposicion_de_capacidad', 'clase', 'disposicion', 'fecha_disp_inscripcion', 'fecha_disp_re_inscripcion'], axis=1, inplace=True)
```

```python
bailables.head()
```

```python
values_count = bailables['estado'].value_counts
values_count
```

```python
bailables = bailables.dropna(subset=['estado'])
```

```python
bailables2 = bailables[bailables['estado'].str.contains('renovado', case=False)]
bailables2
```

```python
unique_values = bailables2['comuna'].unique()
unique_values
```

```python
bailables2['comuna'] = bailables2['comuna'].str.upper()
```

```python
bailables2['comuna'] = bailables2['comuna'].str.replace('COMUNA', '')
```

```python
print(bailables2['comuna'].dtype)
```

```python
# Reemplazar los espacios en blanco con NaN
bailables2['comuna'].replace(' ', np.nan, inplace=True)

# Eliminar las filas con NaN
bailables2.dropna(subset=['comuna'], inplace=True)

# Ahora puedes convertir la columna a int
bailables2['comuna'] = bailables2['comuna'].astype(int)
```

```python
bailables2['comuna'] = bailables2['comuna'].astype(int)
```

```python
bailables2['comuna'] = bailables2['comuna'].astype(str).str.lstrip('0').astype(int)
```

```python
bailables2
```

```python
values_count = bailables2['comuna'].value_counts()
values_count
```

```python
bailables2['capacidad'] = bailables2['capacidad'].str.replace('personas', '')
```

```python
bailables2['capacidad'] = bailables2['capacidad'].str.replace('poersonas', '')
```

```python
bailables2['capacidad'].replace(' ', np.nan, inplace=True)

# Eliminar las filas con NaN
bailables2.dropna(subset=['capacidad'], inplace=True)

# Ahora puedes convertir la columna a int
bailables2['capacidad'] = bailables2['capacidad'].astype(int)
```

```python
bailables2.info()
```

```python
bailables2
```

```python
bailables2.drop(['nombre', 'periodo'], axis=1, inplace=True)
```

```python
bailables2.dropna(subset=['longitud', 'latitud'], inplace=True)
```

```python
bailables2.info()
```

```python
bailables2['barrio'] = bailables2['barrio'].str.title()
bailables2['estado'] = bailables2['estado'].str.title()
```

```python
bailables2
```

```python
barrio_counts = bailables2['barrio'].value_counts()
print(barrio_counts)
```

```python
bailables2['barrio'] = bailables2['barrio'].replace('NuãEz', 'Nuñez')
```

```python
bailables2['barrio'] = bailables2['barrio'].replace(r'Nu\U00F1Ez', 'Nuñez')
```

```python
barrio_counts = bailables2['barrio'].value_counts()
print(barrio_counts)
```

```python
#reordenar columnas 
desired_columns = ['latitud', 'longitud', 'direccion', 'barrio', 'comuna', 'capacidad', 'estado']
bailables3 = bailables2.reindex(columns=desired_columns)
```
